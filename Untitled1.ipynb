```python
import os
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# 1. Dataset Load करना
df = pd.read_csv('Crop_recommendation.csv')
print("Original Dataset Shape:", df.shape)
```

```python
# 2. मौसम आधारित नया डेटा जेनरेट करने का फंक्शन (सॉइल फीचर्स को हटाकर)
def generate_samples(n, temp, humidity, rainfall, label):
    return pd.DataFrame({
        'N':           [0] * n,  # डमी वैल्यू, ट्रेनिंग से पहले ड्राप हो जाएगी
        'P':           [0] * n,
        'K':           [0] * n,
        'temperature': np.random.uniform(temp-2, temp+2, n),
        'humidity':    np.random.uniform(humidity-5, humidity+5, n),
        'ph':          [6.5] * n, 
        'rainfall':    np.random.uniform(rainfall-15, rainfall+15, n),
        'label':       label
    })

# नया डेटा कंबाइन करना (Rainfall को Kaggle स्केल के अनुसार सुधारा गया है)
new_crops = pd.concat([
    generate_samples(100, 20.0, 60.0, 80.0,  'wheat'),       
    generate_samples(100, 30.0, 45.0, 40.0,  'bajra'),       
    generate_samples(100, 26.0, 65.0, 95.0,  'groundnut'),
    generate_samples(100, 25.0, 68.0, 100.0, 'soybean'),     
    generate_samples(100, 28.0, 75.0, 180.0, 'sugarcane'),
    generate_samples(100, 28.0, 50.0, 50.0,  'jowar'),
    generate_samples(100, 18.0, 55.0, 55.0,  'mustard'),
])

df = pd.concat([df, new_crops], ignore_index=True)
print(f"Total crops: {df['label'].nunique()}")
print(f"Total rows after concatenation: {len(df)}")
```

```python
# 3. 🔥 बदलाव: सिर्फ Weather Features (3 कॉलम्स) को सेलेक्ट करना
weather_features = ['temperature', 'humidity', 'rainfall']
X = df[weather_features]
y = df['label']

# हीटमैप सिर्फ वेदर फीचर्स का कोरिलेशन दिखाएगा
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Weather Features Correlation")
plt.show()

# Label Encoding
le = LabelEncoder()
y = le.fit_transform(y)  # फसल के नाम -> नंबर्स में

# Scaling (अब स्केलर सिर्फ इन 3 कॉलम्स को समझता है)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# 4. Model Training (Balanced Class Weight के साथ ताकि नए क्रॉप्स को पूरा वेट मिले)
rf = RandomForestClassifier(n_estimators=150, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
```

```python
# 5. Model Evaluation (Random Forest)
pred = rf.predict(X_test)
print(f"\nRandom Forest Accuracy: {accuracy_score(y_test, pred):.4f}")
print(classification_report(y_test, pred, target_names=le.classes_))
```

```python
# Confusion Matrix Graphical Plot
cm = confusion_matrix(y_test, pred)
fig, ax = plt.subplots(figsize=(15,15))
disp = ConfusionMatrixDisplay(cm, display_labels=le.classes_)
disp.plot(ax=ax, cmap='Blues')
plt.title("Weather-Only Random Forest Confusion Matrix")
plt.show()
```

```python
# Cross Validation
cv_scores = cross_val_score(rf, X_scaled, y, cv=5)
print(f"CV Accuracy: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")

# Feature Importance Graph
feat_imp = pd.Series(rf.feature_importances_, index=weather_features)
feat_imp.sort_values().plot(kind='barh', title='Weather Feature Importance')
plt.show()
```

```python
# 6. मैन्युअल इनपुट टेस्टिंग (सिर्फ 3 वेदर वैल्यूज के साथ)
# फॉर्मेट: [[temperature, humidity, rainfall]]
sample = pd.DataFrame([[20.8, 82.0, 202.9]], columns=weather_features)

sample_scaled = scaler.transform(sample)
pred_sample = rf.predict(sample_scaled)
prob_sample = rf.predict_proba(sample_scaled)

print(f"\nRecommended Crop: {le.inverse_transform(pred_sample)[0].upper()}")
print(f"Confidence: {prob_sample.max()*100:.2f}%")

print("\nTop 3 Contenders:")
top3_idx = prob_sample[0].argsort()[-3:][::-1]
for i in top3_idx:
    print(f" - {le.classes_[i]}: {prob_sample[0][i]*100:.2f}%")
```

```python
# 7. नई पीकल (.pkl) फाइलें सेव करना
with open('crop_model.pkl', 'wb') as f:
    pickle.dump(rf, f)

with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)
```

```python
# 8. SVM मॉडल कम्पेरिजन (वेदर फीचर्स के साथ)
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)

results = pd.DataFrame({
    'Model': ['Random Forest', 'SVM (RBF)'],
    'Train Accuracy': [
        accuracy_score(y_train, rf.predict(X_train))*100,
        accuracy_score(y_train, svm.predict(X_train))*100
    ],
    'Test Accuracy': [
        accuracy_score(y_test, rf.predict(X_test))*100,
        accuracy_score(y_test, svm.predict(X_test))*100
    ]
})
print("\nModel Comparison Results:")
print(results.round(2))
```

```python
# 9. Learning Curve Plot
train_sizes, train_scores, val_scores = learning_curve(
    rf, X_scaled, y, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy'
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_scores.mean(axis=1)*100, 'o-', color='blue', label='Training Accuracy')
plt.plot(train_sizes, val_scores.mean(axis=1)*100, 'o-', color='green', label='Validation Accuracy')
plt.title('Learning Curve - Weather Only Model')
plt.xlabel('Training Samples')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)
plt.show()

print(f"\nTotal crops in model: {len(le.classes_)}")
print(le.classes_)
```
