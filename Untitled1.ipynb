```python
import os
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import (
    train_test_split,
    cross_val_score,
    learning_curve,
    StratifiedKFold
)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# 1. Dataset Load करना
df = pd.read_csv('Crop_recommendation.csv')
print("Original Dataset Shape:", df.shape)
```

```python
def generate_samples(
    n,
    N_range,
    P_range,
    K_range,
    temp_range,
    humidity_range,
    ph_range,
    rainfall_range,
    label
):
    return pd.DataFrame({
        "N": np.random.randint(N_range[0], N_range[1] + 1, n),
        "P": np.random.randint(P_range[0], P_range[1] + 1, n),
        "K": np.random.randint(K_range[0], K_range[1] + 1, n),

        "temperature": np.random.uniform(temp_range[0], temp_range[1], n),
        "humidity": np.random.uniform(humidity_range[0], humidity_range[1], n),
        "ph": np.random.uniform(ph_range[0], ph_range[1], n),
        "rainfall": np.random.uniform(rainfall_range[0], rainfall_range[1], n),

        "label": label
    })

new_crops = pd.concat([

    generate_samples(
        150,
        (80,120),
        (40,70),
        (35,60),
        (18,25),
        (55,70),
        (6.0,7.5),
        (60,100),
        "wheat"
    ),

    generate_samples(
        150,
        (25,45),
        (50,80),
        (25,45),
        (24,30),
        (60,75),
        (6.0,7.5),
        (70,110),
        "soybean"
    ),

    generate_samples(
        150,
        (30,60),
        (20,40),
        (20,35),
        (28,35),
        (35,55),
        (6.0,7.5),
        (25,60),
        "bajra"
    ),

    generate_samples(
        150,
        (80,120),
        (30,60),
        (60,100),
        (25,35),
        (65,85),
        (6.0,7.5),
        (120,220),
        "sugarcane"
    ),

    generate_samples(
        150,
        (20,45),
        (25,45),
        (25,45),
        (25,32),
        (55,75),
        (6.0,7.5),
        (60,100),
        "groundnut"
    ),

    generate_samples(
        150,
        (40,70),
        (20,45),
        (20,40),
        (24,32),
        (40,60),
        (6.0,7.5),
        (30,70),
        "jowar"
    ),

    generate_samples(
        150,
        (30,60),
        (20,40),
        (20,35),
        (15,24),
        (45,65),
        (6.0,7.5),
        (30,70),
        "mustard"
    )

], ignore_index=True)

df = pd.concat([df, new_crops], ignore_index=True)

print("Updated Dataset Shape:", df.shape)
print("Total Crops:", df['label'].nunique())
```

```python
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
X = df[features]
y = df['label']

sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Weather Features Correlation")
plt.show()

le = LabelEncoder()
y = le.fit_transform(y) 


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.25,
    stratify=y,
    random_state=42
)
```

```python
# 4. Model Training
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_split=8,
    min_samples_leaf=4,
    max_features='sqrt',
    bootstrap=True,
    class_weight='balanced',
    random_state=42
)

rf.fit(X_train, y_train)
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    rf,
    X_scaled,
    y,
    cv=cv
)

print("\nCross Validation Accuracy:")
print(scores)
print("Average:", scores.mean())
```

```python
# 5. Model Evaluation (Random Forest)
pred = rf.predict(X_test)
print(f"\nRandom Forest Accuracy: {accuracy_score(y_test, pred):.4f}")
print(classification_report(y_test, pred, target_names=le.classes_))
```

```python
# Confusion Matrix Graphical Plot
cm = confusion_matrix(y_test, pred)
fig, ax = plt.subplots(figsize=(15,15))
disp = ConfusionMatrixDisplay(cm, display_labels=le.classes_)
disp.plot(ax=ax, cmap='Blues')
plt.title("Weather-Only Random Forest Confusion Matrix")
plt.show()
```

```python
# Cross Validation
cv_scores = cross_val_score(rf, X_scaled, y, cv=5)
print(f"CV Accuracy: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")

# Feature Importance Graph
feat_imp = pd.Series(rf.feature_importances_, index=features)
feat_imp.sort_values().plot(kind='barh', title='Weather Feature Importance')
plt.show()
```

```python

sample = pd.DataFrame([[
    90,
    42,
    43,
    20.8,
    82,
    6.5,
    202.9
]], columns=features)

sample_scaled = scaler.transform(sample)
pred_sample = rf.predict(sample_scaled)
prob_sample = rf.predict_proba(sample_scaled)

print(f"\nRecommended Crop: {le.inverse_transform(pred_sample)[0].upper()}")
print(f"Confidence: {prob_sample.max()*100:.2f}%")

print("\nTop 3 Contenders:")

top3_idx = prob_sample[0].argsort()[-3:][::-1]
for i in top3_idx:
    print(f" - {le.classes_[i]}: {prob_sample[0][i]*100:.2f}%")
```

```python

with open('crop_model.pkl', 'wb') as f:
    pickle.dump(rf, f)

with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)
```

```python
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)

results = pd.DataFrame({
    'Model': ['Random Forest', 'SVM (RBF)'],
    'Train Accuracy': [
        accuracy_score(y_train, rf.predict(X_train))*100,
        accuracy_score(y_train, svm.predict(X_train))*100
    ],
    'Test Accuracy': [
        accuracy_score(y_test, rf.predict(X_test))*100,
        accuracy_score(y_test, svm.predict(X_test))*100
    ]
})
print("\nModel Comparison Results:")
print(results.round(2))
```

```python
train_sizes, train_scores, val_scores = learning_curve(
    rf, X_scaled, y, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy'
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_scores.mean(axis=1)*100, 'o-', color='blue', label='Training Accuracy')
plt.plot(train_sizes, val_scores.mean(axis=1)*100, 'o-', color='green', label='Validation Accuracy')
plt.title('Learning Curve - Weather Only Model')
plt.xlabel('Training Samples')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)
plt.show()

print(f"\nTotal crops in model: {len(le.classes_)}")
print(le.classes_)
```
